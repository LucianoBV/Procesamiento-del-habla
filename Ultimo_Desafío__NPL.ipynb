###**Extracción de texto:** Se utiliza la librería pdfplumber para extraer el texto relevante de los PDFs, omitiendo secciones como el índice, prólogo, etc.

```python
# Función para determinar si una página a la que se extraerá el texto
def pagina_a_procesar(texto_pagina):
    # Palabras clave para omitir ciertas secciones como índice, prólogo, anexos, etc.
    palabras_omitir = ['Índice', 'Prólogo', 'Anexo', 'Imágenes', 'Agradecimientos', 'Dedicatoria', 'http',]

    # Revisar si la página contiene alguna palabra que indique que se excluirá
    for palabra in palabras_omitir:
        if palabra.lower() in texto_pagina.lower():
            return False
    return True

# Función para extraer el texto importante del PDF
def extraer_texto(Martin):
    texto_relevante = ''

    with pdfplumber.open(Martin) as pdf:
        pagina_inicio=0
        pagina_fin=len(pdf.pages)
        # Iterar sobre las páginas en el rango seleccionado
        for pagina_numero in range(pagina_inicio, pagina_fin):
            pagina = pdf.pages[pagina_numero]
            texto_pagina = pagina.extract_text()

            # Si la página no tiene texto o es irrelevante, omitirla
            if texto_pagina and pagina_a_procesar(texto_pagina):
                texto_relevante += texto_pagina + '\n'

    return texto_relevante


 # Extraer el texto relevante del rango de páginas
texto_Martin = extraer_texto(Martin)
# Mostrar el texto extraído
print("Texto relevante extraído:")
print(texto_Martin)
```

### **Preprocesamiento del texto:**
Se realiza un preprocesamiento del texto utilizando la librería spacy.

```python
import spacy
```

```python
nlp = spacy.load('es_core_news_sm')  # diccionario en español
```

```python
def eliminar_palabras_vacias(texto):
    """
    Esta función elimina las palabras vacías (stop words) de un texto.

    Args:
        texto: El texto del cual se eliminarán las palabras vacías.

    Returns:
        El texto sin palabras vacías.
    """
    texto = str(texto)
    documento = nlp(texto)
    tokens_sin_palabras_vacias = [token.text for token in documento if not token.is_stop]
    return " ".join(tokens_sin_palabras_vacias)

# Aplica la función al corpus
corpus_sin_palabras_vacias = [eliminar_palabras_vacias(texto) for texto in corpus]
```

### **Tokenización:**
Dividimos el texto en unidades individuales llamadas tokens para la preparación del análisis y la eliminación de ruido.

```python
for text in corpus_sin_palabras_vacias:

    text = str(text)
    doc = nlp(text)
    print('Tipo de dato: ' + str(type(doc)))
    print([w.text for w in doc])
```

## Vectorización:
Se utiliza TF-IDF (TfidfVectorizer) para convertir el texto en representaciones numéricas. TF-IDF calcula la importancia de cada palabra en un documento en relación al corpus. Esto permite que el modelo "entienda" el contenido del texto de forma numérica.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline

# Define las etiquetas para cada obra: 0 para Fausto, 1 para Martín Fierro
etiquetas = np.array([0, 1])

# Vectorización: Convertimos el texto en representaciones numéricas.
vectorizador = TfidfVectorizer()
# TF-IDF calcula la importancia de cada palabra en un documento en relación al corpus.
vectores_tfidf = vectorizador.fit_transform(corpus_sin_palabras_vacias)

```

## Entrenamiento del modelo:

Se utiliza un modelo de clasificación Naive Bayes (MultinomialNB) que se entrena con los vectores TF-IDF y las etiquetas correspondientes (0 para Fausto, 1 para Martín Fierro). Esto significa que el modelo aprende a diferenciar entre los dos textos basándose en las características de las palabras.




```python
# Modelo: Entrenamos un modelo de clasificación (Naive Bayes)
modelo = MultinomialNB()
modelo.fit(vectores_tfidf, etiquetas)
```

### **Clasificación:**
Se define la función clasificar_texto que recibe un nuevo texto, lo preprocesa, lo vectoriza y utiliza el modelo entrenado para predecir a qué obra pertenece.

```python
def clasificar_texto(texto):


    # Preprocesamiento: se limpia y tokeniza
    texto_procesado = eliminar_palabras_vacias(texto)

    # Vectorización: se convierte a formato numérico con TF-IDF
    vector_texto = vectorizador.transform([texto_procesado])

    # Clasificación: se usa el modelo entrenado para predecir
    prediccion = modelo.predict(vector_texto)[0]

    # Se retorna el nombre de la obra según la predicción
    if prediccion == 0:
        return "Fausto"
    else:
        return "Martín Fierro"

# Ejemplo de uso:
nuevo_texto = "En un overo rosilla, zaino cabos negros"
obra_predicha = clasificar_texto(nuevo_texto)
print(f"El texto pertenece a: {obra_predicha}")
nuevo_texto = "En este punto el cantor buscó un porrón pa consuelo"
obra_predicha = clasificar_texto(nuevo_texto)
print(f"El texto pertenece a: {obra_predicha}")
nuevo_texto = "Corcobió la rubiecita Pero al fin se sosegó"
obra_predicha = clasificar_texto(nuevo_texto)
print(f"El texto pertenece a: {obra_predicha}")

```
